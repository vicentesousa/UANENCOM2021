# Caso de estudo 2: Autonomous Underwater Vehicles

## Hands-on 1: comparação de AUVs

### Observação:
* Nesse caso de estudo foi necessário realizar alterações no módulo AUV, segue o patch com as alterações:
* Versão [ns-3.31](https://www.nsnam.org/releases/ns-allinone-3.31.tar.bz2) e [patch](https://github.com/danielrluna/auv-ns3/blob/auv-ns3-31-CE02.patch);


### Cenário:
* Protocolo MAC: Aloha;
* Modelo de AUV: Glider, Remus e Typhoon;
* Modelo de propagação: Ideal - [UanPropModelIdeal](https://www.nsnam.org/doxygen/classns3_1_1_uan_prop_model_ideal.html);
* Modelo de ruído: Padrão - [UanNoiseModelDefault](https://www.nsnam.org/doxygen/classns3_1_1_uan_noise_model_default.html);
* Distância horizontal entre o *Gateway (Sink)* e a posição inicial do AUV: 10m, 1000m, 2000m, 3000m, 4000m e 5000m;
* Tempo de simulação: 1 dia;
* JOBS: 100




### Objetivos:
&emsp; Baseando-se no cenário do exemplo **_uan-auv-ipv4.cc_**, em que foi utilizado apenas um modelo de veículo autônomo (AUV), deseja-se replicar para outros dois modelos de AUV e concluir qual o mais adequado para o cenário em questão. A posição inicial dos AUVs será com uma profundidade de 100m, partindo desse ponto inicial, os veículos irão emergir até a superfície, com uma distância horizontal de 5m a mais de sua posição de partida. Feito isso, irão submergir voltando para os 100m de profundidade e com acréscimo de mais 5m em sua distância horizontal em relação à posição anterior. Já o Gateway, estará localizado na superfície e terá sua distância horizontal em relação ao AUV definida pela simulação, de acordo com os valores apresentados acima. 


### Resultados:
&emsp;A análise de desempenho, no cenário base apresentado acima, foi realizada para três modelos de AUV:

* Glider;
* Remus;
* Typhoon.

&emsp;Em que foi necessário realizar modificações no código base para tornar possível a utilização dos dois novos modelos (Remus e Typhoon).

&emsp;Dessa forma, foram gerados os seguintes plots: **Throughput vs. Distância**, **PLR vs. Distância** e **_Energy Consumed_ vs. Distância (navegação e modem)**. 


### &emsp;*Taxa Útil*:



&emsp;No gráfico de Taxa útil x Distância Horizontal é perceptível que os 3 modelos de AUV obtiveram mesmo desempenho, em que de 10m até 4.000m houve uma leve queda de 13,98 bps para 13,85 bps. Ao chegar na distância de 5.000m, não há mais comunicação entre o modem dos AUVs e a sink, portanto a taxa útil vai para zero. 

### &emsp;*PLR*:



&emsp;No gráfico da PLR, observa-se que os AUVs obtiveram resultados semelhantes, apenas o Typhoon ficou um pouco acima dos outros modelos, mas essa diferença não chega a atingir 1%. De 10m até 4.000m a PLR para o Glider e Remus começou em 27,28% e decaiu lentamente até 27,06%, já o Typhoon saiu de 28,08% e chegou em 27,70%. Ao atingir 5.000m, em que há perda de conexão, tem-se 100% de PLR para todos os AUVs.


### &emsp;*Energia Consumida* (Navegação + Modem):



&emsp;No gráfico de consumo de energia proveniente da navegação e do modem do AUV pode-se perceber que o modelo que mais consumiu energia foi o Remus, seguido do Glider e o de menor consumo foi o Typhoon. De 10m até 4.000m todos os modelos apresentam decaimento lento, em que o Remus parte de um consumo de 1,074MJ até 1,060MJ, o Glider de 1,046MJ até 1,032MJ e o Typhoon de 1,031MJ até 1,023MJ. Vale destacar que em 5.000m há uma queda brusca no consumo de energia para todos os modelos, o que corrobora é o fato da PLR atingir 100%, fazendo com que o modem do AUV permaneça mais tempo em modos diferentes do TX, dessa forma resta o consumo de navegação e o consumo reduzido do modem. Apesar dos pontos inicial e final serem os mesmos para todos os AUVs, cada modelo possui suas limitações e vai performar sua trajetória de acordo com o possível, por isso se tem uma diferença no consumo de energia para esse cenário (em determinado cenário é possível que tenham desempenhos mais semelhantes ou até iguais).

# Simulation Campaings (creation, run and plot)

Our processo to create, run and plot a simulation campaign is completely based on **yaml** files. In this file, a set of parameters is definied as well as the target parameter to a **parametric evaluation**. Such parametric evaluation allows for the analyses of the LoRaWAN metrics in function of the defined range of the target parameter.

The **yaml** file will be used by to Python functions: the **plotScript.py**, which generates all auxiliary files to run simulations; and **plotScript.py**, which plots the simulation results.

Herein, we provide two campaigns we used to perform the proof-of-concept simulations of our ADR proposal. They are:

- **numerologyBwp1** campaign with the following parameters:
   - numerologyBwp1 (target parameter): 1, 2, 3, 4;
   - bandwidthBwp1: 20e6;

- **bandwidthBwp1** campaign with the following parameters:
   - bandwidthBwp1 (target parameter): 10e6, 20e6, 40e6, 80e6, 100e6;
   - numerologyBwp1: 1;

**Attention**: maybe you need to specify the **ns3_path** (with our complete path) to run simulation in our local machine.

### Benchmarking Campaign


````python
import numpy as np

band1 = np.arange(100e6,1e9,12e6)
totalbandwidth = 1e9
band2 = totalbandwidth - band1
freq1 = 39.5e9
freq2 = 40e9

for i in range(len(band1)):
    if band1[i]+band2[i] <= totalbandwidth:
        rcc = band2[i]/band1[i]
        rcc = round(rcc,2)
        #print(rcc)
        if rcc <= 0.51 and rcc >= 0.49:
            print("Rcc: "+ str(rcc) +
                  " freq1: " + str(freq1) +
                  " band1: " + str(band1[i]) + 
                  " freq2: " + str(freq1+band1[i]) +
                  " band2: " + str(band2[i]) +
                  " total: " + str(band1[i]+band2[i]))
        if rcc <= 0.26 and rcc >= 0.25:
            print("Rcc: "+ str(rcc) +
                  " freq1: " + str(freq1) +
                  " band1: " + str(band1[i]) + 
                  " freq2: " + str(freq1+band1[i]) +
                  " band2: " + str(band2[i]) +
                  " total: " + str(band1[i]+band2[i]))
        if rcc <= 0.13 and rcc >= 0.12:
            print("Rcc: "+ str(rcc) +
                  " freq1: " + str(freq1) +
                  " band1: " + str(band1[i]) + 
                  " freq2: " + str(freq1+band1[i]) +
                  " band2: " + str(band2[i]) +
                  " total: " + str(band1[i]+band2[i]))
````

### Campaign 3:

````python
import numpy as np

band1 = np.arange(100e6,1e9,12e6)
totalbandwidth = 1e9
band2 = totalbandwidth - band1
freq1 = 39.5e9
freq2 = 40e9

for i in range(len(band1)):
    if band1[i]+band2[i] <= totalbandwidth:
        rcc = band2[i]/band1[i]
        rcc = round(rcc,2)
        #print(rcc)
        if rcc <= 0.51 and rcc >= 0.49:
            print("Rcc: "+ str(rcc) +
                  " freq1: " + str(freq1) +
                  " band1: " + str(band1[i]) + 
                  " freq2: " + str(freq1+band1[i]) +
                  " band2: " + str(band2[i]) +
                  " total: " + str(band1[i]+band2[i]))
        if rcc <= 0.26 and rcc >= 0.25:
            print("Rcc: "+ str(rcc) +
                  " freq1: " + str(freq1) +
                  " band1: " + str(band1[i]) + 
                  " freq2: " + str(freq1+band1[i]) +
                  " band2: " + str(band2[i]) +
                  " total: " + str(band1[i]+band2[i]))
        if rcc <= 0.13 and rcc >= 0.12:
            print("Rcc: "+ str(rcc) +
                  " freq1: " + str(freq1) +
                  " band1: " + str(band1[i]) + 
                  " freq2: " + str(freq1+band1[i]) +
                  " band2: " + str(band2[i]) +
                  " total: " + str(band1[i]+band2[i]))
````

# Caso de estudo 2: Autonomous Underwater Vehicles

## Hands-on 2: Deslocamento dos AUVs 

### Observação:
* Nesse caso de estudo foi necessário realizar alterações no módulo AUV, segue o patch com as alterações:
* Versão [ns-3.31](https://www.nsnam.org/releases/ns-allinone-3.31.tar.bz2) e [patch](https://github.com/danielrluna/auv-ns3/blob/auv-ns3-31-CE02.patch);

### Cenário

* Distância horizontal entre o gateway e o AUV: 1000m;
* Deslocamento vertical: 10,25,50,75,100m;
* Protocolo MAC: Aloha;
* Modelo de propagação: Ideal - [UanPropModelIdeal](https://www.nsnam.org/doxygen/classns3_1_1_uan_prop_model_ideal.html);
* Modelo de ruído: Padrão - [UanNoiseModelDefault](https://www.nsnam.org/doxygen/classns3_1_1_uan_noise_model_default.html);
* Modelo de AUV: Glider, Remus, Typhoon;
* Protocolo de Rede: IPv4;
* Tempo de simulação: 1 dia;
* Jobs: 100



### Objetivos 
 &emsp;Investigar o comportamento dos 3 AUVs com a variação da distância horizontal e vertical. A variação vertical máxima foi definida, pelo AUV mais restritivo, sendo ele o Remus, logo a profundidade começa em 10 metros, e progride até chegar nos 100 metros, para cada uma dessas profundidades o veículo percorre uma horizontal até 1.000m e retorna ao ponto de partida, totalizando 2.000m de deslocamento.

### Resultados
&emsp;O desempenho desse hands-on foi baseado em três indicadores: **Throughput, PLR e Energia Consumida (Navegação + Modem)**. Todas as curvas serão apresentadas em função da profundidade em que os AUVs se encontram.


### &emsp;Taxa útil:



&emsp;No gráfico de Taxa útil x Profundidade é perceptível que os 3 modelos de AUV obtiveram mesmo desempenho, em que de 10m até 100m obtiveram uma taxa útil média de 17,7 bps.

### &emsp;PLR:



&emsp;No gráfico da PLR x Profundidade também fica claro que os 3 modelos de AUV obtiveram mesmo desempenho, no qual de 10m até 100m obtiveram uma taxa de perda de pacotes em torno de 28%.

### &emsp;Energia Consumida (Navegação + Modem):



&emsp; Já no gráfico do consumo de energia proveniente da navegação e do modem do AUV pode-se perceber que o modelo que mais consumiu energia foi o Remus, seguido do Glider e Typhoon que obtiveram resultados semelhantes. De 10m até 100m todos os modelos de AUV tiveram um gasto energético aproximadamente constante, sendo o Remus em torno de 1,427MJ e os outros dois modelos em torno de 1,261MJ.
